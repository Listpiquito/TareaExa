```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def IPR_Standing(Pr,Pb,J,Pwf):
    q = []
    for pwf in Pwf:
        if pwf >= Pb:
            qo = J*(Pr-pwf)
        else:
            qo = (J*(Pr-Pb)+(J*Pb/1.8)*(1.8*(1-pwf/Pb)-0.8*(1-pwf/Pb)**2))
        q.append(qo)

    return np.array(q)
```

```python
def Z_factor(P,T,Eg):

    Tpc = 170.5 + 307.3*Eg
    Ppc = 709.6 - 58.7*Eg

    Ppr = P/Ppc
    Tpr = (T+460)/Tpc

    A = 1.39*np.sqrt(Tpr-0.92) \
        -0.36*Tpr \
        -0.101

    B = (Ppr*(0.62-0.23*Tpr)+Ppr**2*(0.066/(Tpr-0.86)-0.037)+(0.132*Ppr**6)/np.exp(20.723*(Tpr-1)))

    C = 0.132 - 0.32*np.log10(Tpr)

    D = np.exp(0.715-1.128*Tpr+0.42*Tpr**2)

    return A + (1-A)*np.exp(-B) + C*(Ppr**D)
```

```python
def Rs_Standing(P,T,API,Eg):

    X = (0.0125*API-0.00091*T)

    Rs = Eg * ((P/18.2+1.4)*10**X)**1.2048

    return Rs
```

```python
def Bo_Standing(Rs,Eg,Eo,T):

    Bo = (0.9759+0.00012*(Rs*np.sqrt(Eg/Eo)+1.25*T)**1.2)

    return Bo
```

```python
def Uo(API,T,Rs):

    x = 10**(
        (3.0324- 0.02023*API )*T**(-1.163))

    uod = 10**x - 1

    a = 10.715*(Rs+100)**(-0.515)

    b = 5.44*(Rs+150)**(-0.338)

    return a*uod**b
```

```python
def rho_mixture(HL,rhoL,rhoG):

    return (
        HL*rhoL
        +
        (1-HL)*rhoG
    )
```

```python
def friction_factor(Re,e,D):

    if Re < 2000:

        return 64/Re

    return (
        1/
        (
            -2*np.log10(
                e/(3.7*D)
                +
                5.74/(Re**0.9)
            )
        )
    )**2
```

```python
def pressure_gradient(rhoM,Vm,D,f):

    grav = rhoM/144

    fric = (
        f*rhoM*Vm**2
    ) / (
        2*32.2*D*144
    )

    return grav + fric
```

```python
def VLP_BeggsBrill(q,Pwh,L,d,T,API,Eg,Eo,GOR):

    P = Pwh

    dz = 100

    n = int(L/dz)

    for i in range(n):

        Z = Z_factor(P,T,Eg)

        Rs = Rs_Standing(P,T,API,Eg)

        Bo = Bo_Standing(Rs,Eg,Eo,T)

        muo = Uo(API,T,Rs)

        rhoL = 50
        rhoG = 3

        HL = 0.7

        rhoM = rho_mixture(HL,rhoL,rhoG)

        Vm = 5

        Re = 100000

        f = friction_factor(Re,0.0006,d)

        grad = pressure_gradient( rhoM,Vm,d,f)

        P += grad*dz

    return P
```
